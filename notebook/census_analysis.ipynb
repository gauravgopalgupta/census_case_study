```python
plt.style.use('ggplot') # Using ggplot2 style visuals 

f, ax = plt.subplots(figsize=(11, 40))

ax.set_facecolor('#fafafa')
ax.set(xlim=(-.05, 200))
plt.ylabel('Variables')
plt.title("Overview Data Set")
ax = sns.boxplot(data = dataset, 
  orient = 'h', 
  palette = 'Set2')
```

```python
#Correlation matrix
corr_mat= dataset.corr()
corr_to_arr = np.array(corr_mat)
corr_to_arr[np.tril_indices_from(corr_to_arr)] = False
fig=plt.gcf()
fig.set_size_inches(25,15)
sns.heatmap(data = corr_mat, mask = corr_to_arr, square=True, annot=True, cbar=True)
```

```python
#Analysis of household condition in rural vs urban area
p = dataset.iloc[:,11:14].hist(figsize = (10,10))
```

```python
analysis1 = dataset.groupby(10)[[12, 13, 14]].sum()
analysis1 = analysis1.drop('Total')
analysis1 = analysis1.T
analysis1
```

```python
fig = plt.figure(figsize = (7,7))
plt.bar(['Good', 'Livable', 'Dilapidated'], analysis1['Rural'], color = 'red')
plt.bar(['Good', 'Livable', 'Dilapidated'], analysis1['Urban'], color = 'blue')
plt.legend(analysis1.columns)
plt.ylabel('count')
plt.title('Household condition : Rural/Urban area wise in Bangalore')
plt.show()
```
